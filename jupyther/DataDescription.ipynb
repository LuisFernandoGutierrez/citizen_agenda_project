```
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
from datetime import datetime
```

```
path_data = "/content/drive/My Drive/CAOBA-Central/P017_Nutresa_CuidadoPorLaVida/P017_Nutresa_CuidadoPorLaVida_Nutresa/Muestra de datos/"

files_data = ['AT 2020 Comercializadoras Anonimizado.csv', 
 'AT 2020 Servicios Anonimizado.csv',  
 'Base de datos CAOBA II anonimizado.csv', 
 'CARGUE ACCIDENTALIDAD COMERCIAL NUTRESA Anonimizados.csv',
 'Empleados anonimizado.csv',
 'Terceros Comercial Anonimizado.csv',
 'Terceros Opperar Informacion general Anonimizado.csv', 
 'terceros Servicios Nutresa Anonimizado.csv']
```

```
now = datetime.now()
file_date = now.strftime("%Y-%m-%d_%H-%M")

df_files = []
profiles = []
for item in files_data:
    print("{}{}".format(path_data, item))
    df = pd.read_csv("{}".format(item), delimiter=";", encoding='utf8')
    print(item)
    title_name = item.replace('.csv','').lower()
    df_files.append(df)
    profile = ProfileReport(df, title='Profiling Report: ' + title_name, html={'style':{'full_width':True}})
    profile.to_file(output_file='report/{}_{}.html'.format(file_date, title_name)) 
    profiles.append(profile)

   
   
```
